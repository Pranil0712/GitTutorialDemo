```python
n=int(input("Enter numbers of rows:"))
for i in range(1,n+1):
    print(" "*(n-i),end=" ")
    for j in range(i-1,0-1,-1):
        if j==i-1 and i!=1:
            print(90-(j+1),end=" ")
        else:
            print((90-j),end=" ")
    print()
```
